```python
df=pd.read_csv("casasboston.csv")
data=pd.DataFrame(df)
print(data)
```

```python
data.head()
#data.info()
#data.shape
```

```python
df = data[["RM", "CRIM", "MEDV", "INDUS", "TOWN", "CHAS"]]
print(data)
```

```python
data = data.rename(columns={
    'TOWN': 'CIUDAD',
    'CRIM': 'INDICE_CRIMEN',
    'INDUS': 'PCT_ZONA_INDUSTRIAL',
    'CHAS': 'RIO_CHARLES',
    'RM': 'N_HABITACIONES_MEDIO',
    'MEDV': 'VALOR_MEDIANO',
    'LSTAT': 'PCT_CLASE_BAJA'
})
#print(data)
print(data.columns)
```

HISTOGRAMA DEL NUMERO MEDIO DE HABITACIONES

```python
data.N_HABITACIONES_MEDIO.plot.hist()
plt.show()
```

**MEJORA DE GRAFICACION**

```python
import matplotlib.pyplot as plt

#COLOR

data.N_HABITACIONES_MEDIO.plot.hist(color='purple', alpha=0.5)

#TITULOS CORRESPONDIENTES

plt.xlabel('Número de habitaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma del número de habitaciones medio')

# CUADRICULA
plt.grid(axis='y', alpha=0.75)

plt.show()
```

GRAFICO DE DISPERSION

```python
import matplotlib.pyplot as plt

# GRAFICO DISPERSION
plt.scatter(data['VALOR_MEDIANO'], data['INDICE_CRIMEN'], alpha=0.5)

#TITULOS CORRESPONDIENTES
plt.xlabel('Valor mediano de las viviendas')
plt.ylabel('Índice de criminalidad')
plt.title('Relación entre el valor mediano de las viviendas y el índice de criminalidad')


plt.show()
```

```python
valor_por_ciudad = data.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
print(valor_por_ciudad)

#ENSENA LOS PRIMEROS 10
valor_por_ciudad.head(10).plot.barh()
plt.show()
```
